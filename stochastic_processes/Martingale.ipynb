```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulate_random_walk(steps, num_simulations, X0=0):
    simulations = np.zeros((steps, num_simulations))
    simulations[0, :] = X0
    for i in range(num_simulations):
        for j in range(1, steps):
            step = np.random.choice([-1, 1])
            simulations[j, i] = simulations[j-1, i] + step
    return simulations

def plot_simulations(simulations):
    plt.figure(figsize=(10, 6))
    for i in range(simulations.shape[1]):
        plt.plot(simulations[:, i])
    plt.title('Random Walk Simulations')
    plt.xlabel('Steps')
    plt.ylabel('Position')
    plt.axhline(y=simulations[0, 0], linestyle='-', color='black')
    plt.savefig(f'random_walk_simulations_{simulations.shape[0]}.png')
    plt.show()
```

```python
sample_1 = simulate_random_walk(steps=100, num_simulations=1, X0=0)
plot_simulations(sample_1)
```

```python
sample_2 = simulate_random_walk(steps=1000, num_simulations=1, X0=0)
plot_simulations(sample_2)
```

```python
sample_3 = simulate_random_walk(steps=10000, num_simulations=10, X0=0)
plot_simulations(sample_3)
```

```python
def plot_simulations_bands(simulations):
    n_steps = simulations.shape[0]
    t = np.arange(n_steps)

    std_t = np.sqrt(t)
    plt.figure(figsize=(10,6))
    plt.fill_between(t, -3*std_t, 3*std_t, alpha=0.15, color='red', label='3')
    plt.fill_between(t, -2*std_t, 2*std_t, alpha=0.2, color='red', label='2')
    plt.fill_between(t, -1*std_t, 1*std_t, alpha=0.3, color='red', label='1')

    for i in range(simulations.shape[1]):
        plt.plot(simulations[:, i])

    plt.axhline(0, color='black', linewidth=0.5)
    plt.xlabel('Steps')
    plt.ylabel('Position')
    plt.legend()
    plt.title('Random Walk with bands')
    plt.savefig('random_walk_with_bands.png')
    plt.show()
```

```python
plot_simulations_bands(sample_3)
```
